```python
import sqlite3
import pandas as pd

# Ruta a tu base de datos
db_path = "C:/Users/Nico/Desktop/DATA SCIENCE/PP- VOLUNTAREADO/chivas-ml/data/external/chivas_dw.sqlite"

# Conectar a la base
conn = sqlite3.connect(db_path)

# Leer la vista limpia
df = pd.read_sql("SELECT * FROM ml_dataset_injury14_clean;", conn)

# Cerrar conexión
conn.close()

print(f"✅ Dataset cargado correctamente con {df.shape[0]} filas y {df.shape[1]} columnas.")
df.head()
```

```python
# Info general del dataset
df.info()

# Revisión rápida de nulos
df.isna().sum()

# Estadísticas descriptivas
df.describe().T
```

```python
# Etiqueta
y = df["y_injury14"]

# Variables predictoras (excluimos columnas de id y fechas, y ACWR_end que se repite con CT_Ratio)
X = df.drop(columns=["y_injury14", "id_jugador", "Microciclo_Num", "fref","ACWR_end"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Train:", X_train.shape, " Test:", X_test.shape)
print("Balance entrenamiento:\n", y_train.value_counts(normalize=True))
```

```python
X_train.isna().sum().sort_values(ascending=False).head(15)
```

```python
#Como tengo algunos valores NAN, los vamos a reemplazar por la media así podemos aplicar smote:

from sklearn.impute import SimpleImputer
import pandas as pd

# Crear imputador robusto usando la mediana
imputer = SimpleImputer(strategy='median')

# Ajustar el imputador solo con el conjunto de entrenamiento
X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)

# Aplicar el mismo imputador al conjunto de prueba
X_test_imputed = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)

# Confirmar que no queden NaN
print("🔍 NaN restantes en train:", X_train_imputed.isna().sum().sum())
print("🔍 NaN restantes en test:", X_test_imputed.isna().sum().sum())

```

```python
#Vamors a utilizar SMOTE para crear valores positivos sintéticos para que haya un equilibrio 
# #en nuestros datos (de Test):
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train_imputed, y_train)

print("Después de SMOTE:")
print(y_res.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Repartimos nuevamente el dataset balanceado
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

# Escalador estándar
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test_res)

print("✅ Escalado completo. Datos listos para modelar.")
```

```python
#Vamos a implementar un modelo Random Fores ya que es ideal para el tipo de dataset que tenemos:
#Tolera datasets chicos.
#No requiere normalización estricta.
#Captura relaciones no lineales entre variables físicas y lesiones.
#Da importancia a cada variable (feature importance).

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Modelo base
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)
rf.fit(X_train_res, y_train_res)

# Predicciones
y_pred = rf.predict(X_test_res)

# Evaluación
print("📊 Matriz de confusión:")
print(confusion_matrix(y_test_res, y_pred))

print("\n🔎 Reporte de clasificación:")
print(classification_report(y_test_res, y_pred))
```

### 🧾 Conclusión del modelo base – Random Forest

El modelo de clasificación de riesgo de lesión desarrollado con Random Forest alcanzó un rendimiento muy sólido considerando el tamaño limitado y el desbalance inicial del conjunto de datos.
Tras aplicar SMOTE para equilibrar las clases (lesión / no lesión), los resultados del modelo fueron los siguientes:

Exactitud (Accuracy): 0.90

Precisión (Precision, clase lesión): 0.89

Sensibilidad (Recall, clase lesión): 0.93

F1-Score (clase lesión): 0.91

Estos valores indican que el modelo logra detectar la gran mayoría de las lesiones reales (93 %), manteniendo una tasa de falsos positivos razonable (alrededor del 11 %).
Esto significa que el sistema tiende a prevenir más que omitir, lo cual es ideal en contextos de prevención deportiva, donde es preferible anticipar una posible lesión antes que dejar pasar un caso de riesgo.

Además, el modelo mantiene un buen equilibrio entre ambas clases, sin sesgos marcados hacia la categoría “sin lesión”.
En conjunto, estos resultados demuestran que el enfoque actual es confiable y útil para identificar microciclos o jugadores con alta probabilidad de lesión, pudiendo servir como una herramienta de apoyo para la toma de decisiones del preparador físico.


```python
import pandas as pd
import matplotlib.pyplot as plt

importances = pd.Series(rf.feature_importances_, index=X_res.columns)
importances = importances.sort_values(ascending=True)

plt.figure(figsize=(8,6))
importances.tail(10).plot(kind='barh')
plt.title("🔝 Variables más influyentes en riesgo de lesión")
plt.xlabel("Importancia")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced']
}

grid_search = GridSearchCV(
    rf,
    param_grid,
    cv=3,  # menos folds = menos tiempo
    scoring='recall',  # o 'f1' si querés balance
    n_jobs=-1,  # usa todos los núcleos disponibles
    verbose=2
)

grid_search.fit(X_train_imputed, y_train)

print("✅ Mejor combinación de parámetros:")
print(grid_search.best_params_)
```

```python
best_rf = grid_search.best_estimator_
best_rf.fit(X_res, y_res)

y_pred_test = best_rf.predict(X_test_imputed)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("📊 Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_test))

print("\n📋 Reporte de clasificación:")
print(classification_report(y_test, y_pred_test))
```

### 🧩 Conclusión del modelo injury_prediction_ml

El presente modelo tuvo como objetivo predecir la probabilidad de lesión dentro de los 14 días posteriores a cada microciclo, utilizando las métricas físicas y de carga registradas en la base de datos.
Si bien el flujo de trabajo implementado fue exitoso —incluyendo la imputación de valores faltantes, balanceo con SMOTE, y optimización mediante GridSearchCV— los resultados finales mostraron limitaciones significativas asociadas al tamaño y equilibrio del dataset.

Los principales hallazgos fueron:

El modelo alcanzó una exactitud global (accuracy) de ~0.84, con buen desempeño para la clase “no lesión”,
pero bajo recall (0.25) en la clase “lesión”, indicando dificultad para detectar los casos positivos reales.

El conjunto de datos presentaba un fuerte desbalance (33 lesiones vs. 599 no lesiones), lo cual, a pesar de los ajustes con SMOTE, sigue afectando la capacidad de generalización.

La cantidad total de observaciones (632) y de lesiones efectivas es insuficiente para construir un modelo robusto y estable.

Por estas razones, se decidió pausar el desarrollo del modelo injury_prediction_ml en su estado actual y retomar su optimización cuando se disponga de un volumen mayor de datos de lesiones reales.
