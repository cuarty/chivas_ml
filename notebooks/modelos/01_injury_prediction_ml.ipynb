```python
import sqlite3
import pandas as pd

# Ruta a tu base de datos
db_path = "C:/Users/Nico/Desktop/DATA SCIENCE/PP- VOLUNTAREADO/chivas-ml/data/external/chivas_dw.sqlite"

# Conectar a la base
conn = sqlite3.connect(db_path)

# Leer la vista limpia
df = pd.read_sql("SELECT * FROM ml_dataset_injury14_clean;", conn)

# Cerrar conexi√≥n
conn.close()

print(f"‚úÖ Dataset cargado correctamente con {df.shape[0]} filas y {df.shape[1]} columnas.")
df.head()
```

```python
# Info general del dataset
df.info()

# Revisi√≥n r√°pida de nulos
df.isna().sum()

# Estad√≠sticas descriptivas
df.describe().T
```

```python
# Etiqueta
y = df["y_injury14"]

# Variables predictoras (excluimos columnas de id y fechas, y ACWR_end que se repite con CT_Ratio)
X = df.drop(columns=["y_injury14", "id_jugador", "Microciclo_Num", "fref","ACWR_end"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Train:", X_train.shape, " Test:", X_test.shape)
print("Balance entrenamiento:\n", y_train.value_counts(normalize=True))
```

```python
X_train.isna().sum().sort_values(ascending=False).head(15)
```

```python
#Como tengo algunos valores NAN, los vamos a reemplazar por la media as√≠ podemos aplicar smote:

from sklearn.impute import SimpleImputer
import pandas as pd

# Crear imputador robusto usando la mediana
imputer = SimpleImputer(strategy='median')

# Ajustar el imputador solo con el conjunto de entrenamiento
X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)

# Aplicar el mismo imputador al conjunto de prueba
X_test_imputed = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)

# Confirmar que no queden NaN
print("üîç NaN restantes en train:", X_train_imputed.isna().sum().sum())
print("üîç NaN restantes en test:", X_test_imputed.isna().sum().sum())

```

```python
#Vamors a utilizar SMOTE para crear valores positivos sint√©ticos para que haya un equilibrio 
# #en nuestros datos (de Test):
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train_imputed, y_train)

print("Despu√©s de SMOTE:")
print(y_res.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Repartimos nuevamente el dataset balanceado
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

# Escalador est√°ndar
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test_res)

print("‚úÖ Escalado completo. Datos listos para modelar.")
```

```python
#Vamos a implementar un modelo Random Fores ya que es ideal para el tipo de dataset que tenemos:
#Tolera datasets chicos.
#No requiere normalizaci√≥n estricta.
#Captura relaciones no lineales entre variables f√≠sicas y lesiones.
#Da importancia a cada variable (feature importance).

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Modelo base
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)
rf.fit(X_train_res, y_train_res)

# Predicciones
y_pred = rf.predict(X_test_res)

# Evaluaci√≥n
print("üìä Matriz de confusi√≥n:")
print(confusion_matrix(y_test_res, y_pred))

print("\nüîé Reporte de clasificaci√≥n:")
print(classification_report(y_test_res, y_pred))
```

### üßæ Conclusi√≥n del modelo base ‚Äì Random Forest

El modelo de clasificaci√≥n de riesgo de lesi√≥n desarrollado con Random Forest alcanz√≥ un rendimiento muy s√≥lido considerando el tama√±o limitado y el desbalance inicial del conjunto de datos.
Tras aplicar SMOTE para equilibrar las clases (lesi√≥n / no lesi√≥n), los resultados del modelo fueron los siguientes:

Exactitud (Accuracy): 0.90

Precisi√≥n (Precision, clase lesi√≥n): 0.89

Sensibilidad (Recall, clase lesi√≥n): 0.93

F1-Score (clase lesi√≥n): 0.91

Estos valores indican que el modelo logra detectar la gran mayor√≠a de las lesiones reales (93 %), manteniendo una tasa de falsos positivos razonable (alrededor del 11 %).
Esto significa que el sistema tiende a prevenir m√°s que omitir, lo cual es ideal en contextos de prevenci√≥n deportiva, donde es preferible anticipar una posible lesi√≥n antes que dejar pasar un caso de riesgo.

Adem√°s, el modelo mantiene un buen equilibrio entre ambas clases, sin sesgos marcados hacia la categor√≠a ‚Äúsin lesi√≥n‚Äù.
En conjunto, estos resultados demuestran que el enfoque actual es confiable y √∫til para identificar microciclos o jugadores con alta probabilidad de lesi√≥n, pudiendo servir como una herramienta de apoyo para la toma de decisiones del preparador f√≠sico.


```python
import pandas as pd
import matplotlib.pyplot as plt

importances = pd.Series(rf.feature_importances_, index=X_res.columns)
importances = importances.sort_values(ascending=True)

plt.figure(figsize=(8,6))
importances.tail(10).plot(kind='barh')
plt.title("üîù Variables m√°s influyentes en riesgo de lesi√≥n")
plt.xlabel("Importancia")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced']
}

grid_search = GridSearchCV(
    rf,
    param_grid,
    cv=3,  # menos folds = menos tiempo
    scoring='recall',  # o 'f1' si quer√©s balance
    n_jobs=-1,  # usa todos los n√∫cleos disponibles
    verbose=2
)

grid_search.fit(X_train_imputed, y_train)

print("‚úÖ Mejor combinaci√≥n de par√°metros:")
print(grid_search.best_params_)
```

```python
best_rf = grid_search.best_estimator_
best_rf.fit(X_res, y_res)

y_pred_test = best_rf.predict(X_test_imputed)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("üìä Matriz de confusi√≥n:")
print(confusion_matrix(y_test, y_pred_test))

print("\nüìã Reporte de clasificaci√≥n:")
print(classification_report(y_test, y_pred_test))
```
